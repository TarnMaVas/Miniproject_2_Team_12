Копайлот сам розбив 4 функції на 4 класи і в них створив функції для їх тестування.
Також він використав 2 шаблони у двох функціях в перевірках конктретних тест-кейсах.





З ствроенням тестів він справився чудовою

2)

Щодо помилок, то тести, які він згенерував, чудово з цим справились. Можна побачити, що ламається і видає помилку (який рядок), які конкретно функції працюють некоректно.

3)

З покриттям він справився зразу ж, з 1 ж генерації. Що, як по мені, дуже великий плюс і може оптимізувати твою роботу.



4)

По факту, він з першої ж генерації (можливо через те, що я вже працював з ним і знаю, за допомогою яких конкретно промптів він видає найкращий результат) дуже лаконічно підібрав все тест-кейси. Можна, звичайно, пошукати тести, які можна забрати або додати, але це вже не є великою необхідністю, бо ці кейси ітак перевіряють все, що потрібно і працює, і підбирає тести, які показують, що не працює.

Загальний висновок:

Копайлот зараз є одним з найкращих ШІ, який працює з кодом. Якщо ним довго користуватися і знати за яким шаблоном він генерує код і допомагає з написанням, то можна як і покращити свій код, так і згенерувати новий (не ідеальний звісно, но майже завжди працюючий). Є у нього звичайно і великі мінуси, але з ними швидко змирюєшся і перекриваєш ще сильнішими плюсами. Так що, якщо ти хочеш дізнатися, що у тебе не працює, не маючи кінцевих тестів, то він справляється з цим чудово.
