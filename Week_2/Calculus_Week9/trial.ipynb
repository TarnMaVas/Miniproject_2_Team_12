Here you can see the visualization of the time difference between my and ai-optimized code

```python
import time
import numpy as np
import matplotlib.pyplot as plt
import initial_solution
import ai_code

list_lengths = [100000, 200000, 3000000, 4000000, 5000000]

my_calculating_times = []
ai_calculating_times = []

for length in list_lengths:
    points = list(np.random.randint(low=-1000000, high=1000000, size=length))
    poly = lambda x: x**2 + 3*x + 3 # this is just a random one, others could be chosen as well; the result will be the same
    start = time.time()
    initial_solution.find_max_1(poly, points)
    end = time.time()
    my_calculating_times.append(end - start)

    start = time.time()
    ai_code.find_max_1(poly, points)
    end = time.time()
    ai_calculating_times.append(end - start)

plt.plot(list_lengths, my_calculating_times, label='My calculating time')
plt.plot(list_lengths, ai_calculating_times, label='Blackbox calculating time')
plt.xlabel('Number of points to check')
plt.ylabel('Time (seconds)')
plt.title('My and Blackbox calculating time vs. Number of points')
plt.legend()
plt.show()
```

```python
import time
import matplotlib.pyplot as plt
import initial_solution
import ai_code

degrees = [10000, 20000, 50000, 75000, 100000]

my_calculating_times = []
ai_calculating_times = []

for degree in degrees:
    func = lambda n: 2 * (n ** degree + n) / n ** degree # this is just a random one, others could be chosen as well; the result will be the same
    start = time.time()
    initial_solution.compute_limit(func)
    end = time.time()
    my_calculating_times.append(end - start)

    start = time.time()
    ai_code.compute_limit(func)
    end = time.time()
    ai_calculating_times.append(end - start)

plt.plot(degrees, my_calculating_times, label='My computation time')
plt.plot(degrees, ai_calculating_times, label='Blackbox computation time')
plt.xlabel('Degree of sequence')
plt.ylabel('Time (seconds)')
plt.title('My and Blackbox computation time vs. Sequence degree')
plt.legend()
plt.show()
```
