Here you can see the visualization of the time difference between my and ai-optimized code

```python
import time
import random
import string
import matplotlib.pyplot as plt
import initial_solution
import ai_code

text_lengths = [50000, 100000, 200000, 500000, 1000000]

my_encode_times = []
ai_encode_times = []

for length in text_lengths:
    text = ''.join(random.choices(string.ascii_lowercase, k=length))
    offset = random.randint(0, 25)
    start = time.time()
    initial_solution.caesar_encode(text, offset)
    end = time.time()
    my_encode_times.append(end - start)

    start = time.time()
    ai_code.caesar_encode(text, offset)
    end = time.time()
    ai_encode_times.append(end - start)

plt.plot(text_lengths, my_encode_times, label='My encoding time')
plt.plot(text_lengths, ai_encode_times, label='Blackbox encoding time')
plt.xlabel('Text Length')
plt.ylabel('Time (seconds)')
plt.title('My and Blackbox encoding time vs. Text Length')
plt.legend()
plt.show()
```
